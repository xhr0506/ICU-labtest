# PLAGH

```python
import pandas as pd
import numpy as np
```

```python
df_plagh=pd.read_csv('../datasets/plagh_liquid.csv',encoding='gbk')
df_plagh.head(5)
```

```python
#Chinese names of the items
#blood='红细胞'
crystalloid=['氯化钠','生理盐水','乳酸钠林格','葡萄糖']
colloid=['人血白蛋白','琥珀酰明胶','羟乙基淀粉','聚明胶肽','右旋糖酐']
```

```python
#Make flags of the fluid
#df_plagh.loc[df_plagh['ORDER_TEXT'].str.contains(blood),'BLOOD']=1
for item_1 in crystalloid:
    df_plagh.loc[df_plagh['ORDER_TEXT'].str.contains(item_1),'CRYSTALLOID']=1
for item_2 in colloid:
    df_plagh.loc[df_plagh['ORDER_TEXT'].str.contains(item_2),'COLLOID']=1
```

```python
df_plagh.head(10)
```

```python
df_fluid=df_plagh.loc[:,['PATIENT_ID','VISIT_ID','ORDER_TEXT','DOSAGE','DOSAGE_UNITS','CRYSTALLOID','COLLOID']]
df_fluid
```

```python
df_crys_v=df_fluid.loc[(df_fluid['CRYSTALLOID']==1) & (df_fluid['DOSAGE']>=250),['PATIENT_ID','VISIT_ID','DOSAGE']].groupby(['PATIENT_ID','VISIT_ID']).sum().reset_index()
df_crys_v.rename(columns={'DOSAGE':'CRYSTALLOID_V'},inplace=True)
df_fluid=pd.merge(df_fluid,df_crys_v,on=['PATIENT_ID','VISIT_ID'],how='left')
df_fluid
```

```python
df_fluid.sort_values(by=['PATIENT_ID','VISIT_ID'])
```

```python
df_crys_v.loc[df_crys_v['PATIENT_ID']=='S118060109',:]
```

```python
df_fluid.drop(['ORDER_TEXT','DOSAGE','DOSAGE_UNITS'],axis=1,inplace=True)
df_fluid
```

```python
df_fluid['COLLOID'].fillna(0,inplace=True)
```

```python
df_fluid_crystalloid=df_fluid.loc[:,['PATIENT_ID','VISIT_ID','CRYSTALLOID_V']].drop_duplicates()
df_fluid_crystalloid
```

```python
df_fluid_crystalloid['CRYSTALLOID_V'].notnull().sum()
```

```python
df_fluid_colloid=df_fluid.loc[:,['PATIENT_ID','VISIT_ID','COLLOID']].groupby(['PATIENT_ID','VISIT_ID']).sum().reset_index()
df_fluid_colloid
```

```python
df_fluid_colloid['COLLOID'].replace(2,1,inplace=True)
df_fluid_colloid['COLLOID'].sum()
```

```python
df_plagh_blood=pd.read_csv('../datasets/plagh_blood.csv',encoding='gbk')
df_plagh_blood
```

```python
df_blood=df_plagh_blood.loc[:,['PATIENT_ID','VISIT_ID']].drop_duplicates()
df_blood['BLOOD']=1
df_blood
```

```python
df_plagh_ed=pd.read_csv('../datasets/plagh_ed.csv',encoding='gbk')
df_plagh_ed
```

```python
df_ed=df_plagh_ed.loc[:,['PATIENT_ID','VISIT_ID']].drop_duplicates()
df_ed['EMERGENCY']=1
df_ed
```

```python
df_plagh=pd.merge(df_fluid_crystalloid,df_fluid_colloid,on=['PATIENT_ID','VISIT_ID'],how='outer')
df_plagh=pd.merge(df_plagh,df_blood,on=['PATIENT_ID','VISIT_ID'],how='outer')
df_plagh=pd.merge(df_plagh,df_ed,on=['PATIENT_ID','VISIT_ID'],how='outer')
df_plagh
```

```python
df_plagh.sort_values(by=['PATIENT_ID','VISIT_ID'],inplace=True)
for col in df_plagh.columns.values[2:]:
    df_plagh[col].fillna(0,inplace=True)
df_plagh
```

```python
df_plagh.loc[:,'COLLOID':].sum()
```

```python
df_plagh.to_csv('../datasets/PLAGH-data-update202012.csv')
```

# MIMIC

```python
df_mimic_fluid_mv=pd.read_csv('../datasets/mimic_fluid_mv.csv')
df_mimic_fluid_mv
```

```python
df_mimic_crystalloid_mv=df_mimic_fluid_mv.loc[(df_mimic_fluid_mv['f_type']=='crystalloids') & (df_mimic_fluid_mv['amount']>=249),['subject_id','hadm_id','icustay_id','amount']]
df_mimic_crystalloid_mv=df_mimic_crystalloid_mv.groupby(['subject_id','hadm_id','icustay_id']).sum().reset_index()
```

```python
df_mimic_crystalloid_mv.rename(columns={'amount':'crystalloid_v'},inplace=True)
```

```python
df_mimic_colloid_mv=df_mimic_fluid_mv.loc[(df_mimic_fluid_mv['f_type']=='colloids'),['subject_id','hadm_id','icustay_id']].drop_duplicates()
df_mimic_colloid_mv['colloid']=1
df_mimic_colloid_mv
```

```python
df_mimic_fluid_cv=pd.read_csv('../datasets/mimic_fluid_cv.csv')
df_mimic_fluid_cv
```

```python
df_mimic_crystalloid_cv=df_mimic_fluid_cv.loc[(df_mimic_fluid_cv['f_type']=='crystalloids') & (df_mimic_fluid_cv['amount']>=249),['subject_id','hadm_id','icustay_id','amount']]
df_mimic_crystalloid_cv=df_mimic_crystalloid_cv.groupby(['subject_id','hadm_id','icustay_id']).sum().reset_index()
df_mimic_crystalloid_cv.rename(columns={'amount':'crystalloid_v'},inplace=True)
df_mimic_crystalloid_cv
```

```python
df_mimic_colloid_cv=df_mimic_fluid_cv.loc[(df_mimic_fluid_cv['f_type']=='colloids'),['subject_id','hadm_id','icustay_id']].drop_duplicates()
df_mimic_colloid_cv['colloid']=1
df_mimic_colloid_cv
```

```python
df_mimic_colloid=pd.concat([df_mimic_colloid_cv,df_mimic_colloid_mv],ignore_index=True)
df_mimic_crystalloid=pd.concat([df_mimic_crystalloid_cv,df_mimic_crystalloid_mv],ignore_index=True)
```

```python
df_mimic_blood_cv=pd.read_csv('../datasets/mimic_blood_cv.csv')
df_mimic_blood_cv=df_mimic_blood_cv.loc[:,['subject_id','hadm_id','icustay_id']].drop_duplicates()
df_mimic_blood_mv=pd.read_csv('../datasets/mimic_blood_mv.csv')
df_mimic_blood_mv=df_mimic_blood_mv.loc[:,['subject_id','hadm_id','icustay_id']].drop_duplicates()
df_mimic_blood=pd.concat([df_mimic_blood_cv,df_mimic_blood_mv],ignore_index=True)
df_mimic_blood['blood']=1
df_mimic_blood
```

```python
df_mimic_ed=pd.read_csv('../datasets/mimic_ed.csv')
df_mimic_ed=df_mimic_ed.loc[df_mimic_ed['delta_t']=='ED',['subject_id','hadm_id','icustay_id']].drop_duplicates()
df_mimic_ed['emergency']=1
df_mimic_ed
```

```python
df_mimic=pd.merge(df_mimic_crystalloid,df_mimic_colloid,on=['subject_id','hadm_id','icustay_id'],how='outer')
df_mimic=pd.merge(df_mimic,df_mimic_blood,on=['subject_id','hadm_id','icustay_id'],how='outer')
df_mimic=pd.merge(df_mimic,df_mimic_ed,on=['subject_id','hadm_id','icustay_id'],how='outer')
df_mimic
```

```python
df_mimic.sort_values(by=['subject_id','hadm_id','icustay_id'],inplace=True)
for col in df_mimic.columns.values[3:]:
    df_mimic[col].fillna(0,inplace=True)
df_mimic
```

```python
df_mimic.to_csv('../datasets/MIMIC-data-update202012.csv')
```

# eICU

```python
df_eicu_fluid=pd.read_csv('../datasets/eicu_fluid.csv')
df_eicu_fluid.head(20)
```

```python
df_eicu_crystalloid=df_eicu_fluid.loc[(df_eicu_fluid['f_type']=='crystalloids') & (df_eicu_fluid['cellvaluenumeric']>=249),['patientunitstayid','cellvaluenumeric']].groupby(['patientunitstayid']).sum().reset_index()
df_eicu_crystalloid.rename(columns={'cellvaluenumeric':'crystalloid_v'},inplace=True)
```

```python
df_eicu=pd.merge(df_eicu_crystalloid,df_eicu_colloid,on='patientunitstayid',how='outer')
df_eicu=pd.merge(df_eicu,df_eicu_blood,on='patientunitstayid',how='outer')
df_eicu
```

```python
df_eicu.sort_values(by=['patientunitstayid'],inplace=True)
for col in df_eicu.columns.values[1:]:
    df_eicu[col].fillna(0,inplace=True)
df_eicu
```

```python
df_eicu.to_csv('../datasets/eICU-data-update202012.csv')
```

# AUMC

```python
df_id=pd.read_csv('../datasets/AUMC-data.csv')
df_id
```

```python
df_id=df_id.loc[:,['patientid','admissionid']]
df_id
```

```python
df_aumc_rrt=pd.read_csv('../datasets/AUMC_rrt.csv')
df_aumc_rrt
```

```python
df_aumc_rrt['crrt_flag']=1
df_aumc_rrt.drop(['renal_replacement_bool','renal_replacement_max_blood_flow','renal_replacement_modes'],axis=1,inplace=True)
df_aumc_rrt.drop_duplicates(inplace=True)
df_aumc_rrt.sort_values(by='admissionid',inplace=True)
df_aumc_rrt
```

```python
df_aumc_vent=pd.read_csv('../datasets/AUMC_vent.csv')
df_aumc_vent
```

```python
df_aumc_vent['vent_flag']=1
df_aumc_vent.drop(['mechanical_ventilation_bool','mechanical_ventilation_modes'],axis=1,inplace=True)
df_aumc_vent.drop_duplicates(inplace=True)
df_aumc_vent.sort_values(by='admissionid',inplace=True)
df_aumc_vent
```

```python
df_aumc_blood=pd.read_csv('../datasets/AUMC_rbc.csv')
df_aumc_blood
```

```python
df_aumc_blood.info()
```

```python
df_aumc_blood.loc[df_aumc_blood['no_RBC_transfusion']!=0,'blood']=1
df_aumc_blood['blood'].fillna(0,inplace=True)
df_aumc_blood.drop('no_RBC_transfusion',axis=1,inplace=True)
df_aumc_blood
```

```python
df_aumc_ed=pd.read_csv('../datasets/AUMC_ed.csv')
df_aumc_ed
```

```python
df_aumc_ed.loc[df_aumc_ed['origin_ED']==1,'emergency']='ED'
df_aumc_ed.loc[df_aumc_ed['origin_ED']==0,'emergency']='Other'
df_aumc_ed.drop(['origin','origin_ED'],axis=1,inplace=True)
df_aumc_ed
```

```python
df_aumc_fluid=pd.read_csv('../datasets/AUMC_fluid.csv')
df_aumc_fluid
```

```python
df_aumc_colloid=df_aumc_fluid.loc[df_aumc_fluid['ordercategory'].str.contains('Colloid'),:]
df_aumc_colloid['colloid']=1
df_aumc_colloid.drop(['ordercategory','sum_fluidin'],axis=1,inplace=True)
df_aumc_colloid
```

```python
df_aumc_crystalloid=df_aumc_fluid.loc[df_aumc_fluid['ordercategory'].str.contains('Crystalloid'),:]
df_aumc_crystalloid['crystalloid']=1
df_aumc_crystalloid.drop('ordercategory',axis=1,inplace=True)
df_aumc_crystalloid.rename(columns={'sum_fluidin':'crystalloid_v'},inplace=True)
df_aumc_crystalloid
```

```python
np.median(df_aumc_crystalloid['crystalloid_v'])
```

```python
df_aumc=pd.merge(df_id,df_aumc_rrt,on='admissionid',how='left')
df_aumc=pd.merge(df_aumc,df_aumc_vent,on='admissionid',how='left')
df_aumc=pd.merge(df_aumc,df_aumc_blood,on='admissionid',how='left')
df_aumc=pd.merge(df_aumc,df_aumc_crystalloid,on='admissionid',how='left')
df_aumc=pd.merge(df_aumc,df_aumc_colloid,on='admissionid',how='left')
df_aumc=pd.merge(df_aumc,df_aumc_ed,on='admissionid',how='left')
df_aumc
```

```python
for col in df_aumc.columns.values[2:]:
    df_aumc[col].fillna(0,inplace=True)
df_aumc
```

```python
df_aumc.to_csv('../datasets/AUMC-data-updated202012.csv')
```
