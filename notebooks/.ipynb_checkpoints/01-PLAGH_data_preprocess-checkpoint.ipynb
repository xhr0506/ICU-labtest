```python
import pandas as pd
import numpy as np
```

# Cohort

```python
path='F:\\项目\\301icu\\Leo文章\\1226\\cohort_1226.csv'
f=open(path, encoding='gbk',errors='ignore')
cohort_df=pd.read_csv(f)
cohort_df
```

```python
#age>18, Chinese standard
cohort_18=cohort_df.loc[(cohort_df['AGE']>=18) & (cohort_df['LOS_ICU']> 0) & (cohort_df['LOS_ICU'].notnull()),:]
cohort_18.drop(['ROW_ID','STARTTIME','ENDTIME','DOD_HOSP'],axis=1,inplace=True)
cohort_18
```

# lab test results

```python
path='F:\\项目\\301icu\\Leo文章\\1226\\labresult.csv'
f=open(path, encoding='gbk',errors='ignore')
lab_result=pd.read_csv(f)
lab_result
```

```python
lab_result.replace([u'（',u'已复查',u'）','(',')','[',']',u'【',u'】'],['','','','','','','','',''],inplace=True)
lab_result['RESULT']=pd.to_numeric(lab_result['RESULT'],errors='coerce')
```

```python
lab_result.loc[lab_result['REPORT_ITEM_NAME']=='实际碳酸氢根测定',:]
```

```python
lab_result_pivot=pd.pivot_table(lab_result,values='RESULT',index=['PATIENT_ID','VISIT_ID','TEST_NO'],
                            columns='REPORT_ITEM_NAME')
lab_result_pivot
```

```python
lab_result_pivot.reset_index(inplace=True)
lab_result_pivot
```

```python
lab_result_pivot.rename(columns={'游离钙':'ionizedcalcium', 
                                 '实际碳酸氢根测定':'bicarbonate',
                                 '乳酸':'lactate',
                                 '白细胞计数':'wbc',                       
                                 '肌酐':'creatinine',
                                 '葡萄糖':'glucose',
                                 '血小板计数':'platelets',
                                 '血清白蛋白':'albumin',
                                 '血红蛋白测定':'hemoglobin',
                                 '钙':'calcium',
                                 '钠':'sodium',
                                 '钾':'potassium',
                                 '镁':'magnesium'},inplace=True)
lab_result_pivot
```

```python
lab_result_pivot.drop(['TEST_NO','无机磷'],axis=1,inplace=True)
lab_result_pivot
```

```python
grouped=lab_result_pivot.groupby(['PATIENT_ID','VISIT_ID'])
lab_result_min=grouped.min()
lab_result_min.reset_index(inplace=True)
lab_result_min
```

```python
lab_result_max=grouped.max()
lab_result_max.reset_index(inplace=True)
lab_result_max
```

# extend information

```python
extend_df = pd.read_csv('../datasets/admtype_hosplos.csv')
extend_df.drop_duplicates(inplace=True)
extend_df
```

# merge

```python
plagh_df=pd.merge(cohort_18,extend_df,on=['PATIENT_ID','VISIT_ID'],how='left')
```

```python
plagh_df=pd.merge(plagh_df,lab_result_min,on=['PATIENT_ID','VISIT_ID'],how='left')
plagh_df.info()
```

```python
plagh_df=pd.merge(plagh_df,lab_result_max,on=['PATIENT_ID','VISIT_ID'],how='left')
plagh_df.info()
```

# rename the columns

```python
columns_name=[]
for item in plagh_df.columns.values:
    item=item.replace('_x','_min')
    item=item.replace('_y','_max')
    columns_name.append(item.lower())
```

```python
plagh_df.columns=columns_name
```

```python
plagh_df.info()
```

```python
plagh_df.rename(columns={'icu_type':'unittype',
                         'los_icu':'icu_los_hours',
                        'death_flag':'icu_expire_flag',
                        'hospital_los':'hospital_los_hours'},inplace=True)
plagh_df.info()
```
