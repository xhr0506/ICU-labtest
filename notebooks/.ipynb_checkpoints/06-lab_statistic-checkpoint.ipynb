```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
from tqdm import tqdm
from tableone import TableOne
import datetime
import time
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':20})
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')


# Import libraries
import xgboost as xgb
import statsmodels.api as sm
```

```python
data_mimic = pd.read_csv('../datasets/processed data/MIMIC-data.csv')
data_aumc = pd.read_csv('../datasets/processed data/AUMC-data.csv')
data_eicu = pd.read_csv('../datasets/processed data/eCRD-data.csv')
data_plagh = pd.read_csv('../datasets/processed data/PLAGH-data.csv')
data_HJ23 = pd.read_csv('../datasets/processed data/HJ23-data.csv')
```

```python
# data_plagh['lactate_max']=np.nan
# data_eicu['admission_type']=np.nan
# data_HJ23['cardiac_surgery_flag']=np.nan
```

```python
data_mimic_15=data_mimic.loc[(data_mimic['outcome_group']==1) | (data_mimic['outcome_group']==5),
                             ['gender','age_group','unittype','admission_type','cardiac_surgery_flag',
                              'sodium_max','creatinine_max','lactate_max','hemoglobin_min','outcome_group']]
data_mimic_15['database_flag']='MIMIC'

data_aumc_15=data_aumc.loc[(data_aumc['outcome_group']==1) | (data_aumc['outcome_group']==5),
                             ['gender','age_group','unittype','admission_type','cardiac_surgery_flag',
                              'sodium_max','creatinine_max','lactate_max','hemoglobin_min','outcome_group']]
data_aumc_15['database_flag']='AUMC'

data_eicu_15=data_eicu.loc[(data_eicu['outcome_group']==1) | (data_eicu['outcome_group']==5),
                             ['gender','age_group','unittype','admission_type','cardiac_surgery_flag',
                              'sodium_max','creatinine_max','lactate_max','hemoglobin_min','outcome_group']]
data_eicu_15['database_flag']='eICU'

data_plagh_15=data_plagh.loc[(data_plagh['outcome_group']==1) | (data_plagh['outcome_group']==5),
                             ['gender','age_group','unittype','admission_type','cardiac_surgery_flag',
                              'sodium_max','creatinine_max','lactate_max','hemoglobin_min','outcome_group']]
data_plagh_15['database_flag']='PLAGH'


data_HJ23_15=data_HJ23.loc[(data_HJ23['outcome_group']==1) | (data_HJ23['outcome_group']==5),
                             ['gender','age_group','unittype','admission_type','cardiac_surgery_flag',
                              'sodium_max','creatinine_max','lactate_max','hemoglobin_min','outcome_group']]
data_HJ23_15['database_flag']='HJ23'
```

```python
bestvsworst_df=pd.concat([data_mimic_15,data_aumc_15,data_eicu_15,data_plagh_15,data_HJ23_15],ignore_index=True)
bestvsworst_df
```

```python
for column in bestvsworst_df.columns.values:
    print(column)
    print(bestvsworst_df[column].unique())
```

```python
bestvsworst_df['gender'].replace('U',np.nan,inplace=True)
bestvsworst_df['gender'].replace('1',1,inplace=True)
bestvsworst_df['gender'].replace('0',0,inplace=True)
bestvsworst_df['gender'].unique()
```

```python
bestvsworst_df.info()
```

```python
bestvsworst_df['outcome_group'].replace(1,'Best',inplace=True)
bestvsworst_df['outcome_group'].replace(5,'Worst',inplace=True)
```

```python
columns=['gender',
         'age_group',
         'unittype',
         'admission_type',
         'cardiac_surgery_flag',
        'sodium_max',
         'creatinine_max',
         'lactate_max',
         'hemoglobin_min',
         'database_flag']
categorical=['gender',
             'age_group',
             'unittype',
             'admission_type',
             'cardiac_surgery_flag',
             'database_flag']
groupby=['outcome_group']
nonnormal=['sodium_max','creatinine_max','lactate_max','hemoglobin_min']
mytable=TableOne(bestvsworst_df,columns,categorical,groupby,nonnormal,pval=True)
print(mytable)
```

```python
bestvsworst_df['age_group'].replace('(17, 39]',0,inplace=True)
bestvsworst_df['age_group'].replace('(39, 49]',1,inplace=True)
bestvsworst_df['age_group'].replace('(49, 59]',2,inplace=True)
bestvsworst_df['age_group'].replace('(59, 69]',3,inplace=True)
bestvsworst_df['age_group'].replace('(69, 79]',4,inplace=True)
bestvsworst_df['age_group'].replace('(79, 300]',5,inplace=True)

bestvsworst_df['unittype'].replace('Med-Surg ICU','MedSurgICU',inplace=True)
bestvsworst_df['unittype'].replace('CCU-CTICU','CCUCTICU',inplace=True)
bestvsworst_df['unittype'].replace('SICU&MICU','SICUMICU',inplace=True)
```

```python
# encoding
data_type1 = bestvsworst_df
names_dummies = ['age_group', 'unittype', 'admission_type', 'database_flag','gender']
for i in range(len(names_dummies)):
  vars()['dummies_' + names_dummies[i]] = pd.get_dummies(data_type1[names_dummies[i]], prefix=names_dummies[i])
  data_type1 = pd.concat([data_type1, vars()['dummies_' + names_dummies[i]]], axis=1)
data_type1.drop(names_dummies, axis=1, inplace=True)
data_type1
```

```python
data_type1.rename(columns={'gender_0.0':'gender_0',
                           'gender_1.0':'gender_1'},inplace=True)
data_type1['outcome_group'].replace('Worst',1,inplace=True)
data_type1['outcome_group'].replace('Best',0,inplace=True)
```

```python
for column in data_type1.drop('outcome_group',axis=1).columns.values:
    model = sm.GLM.from_formula("outcome_group ~ {}".format(column), family=sm.families.Binomial(), data=data_type1)
    result = model.fit()
    print(result.summary())
```

```python
data_type1['sodium_max']=data_type1['sodium_max'].fillna(np.nanmedian(data_type1['sodium_max']))
data_type1['creatinine_max']=data_type1['creatinine_max'].fillna(np.nanmedian(data_type1['creatinine_max']))
data_type1['lactate_max']=data_type1['lactate_max'].fillna(np.nanmedian(data_type1['lactate_max']))
data_type1['hemoglobin_min']=data_type1['hemoglobin_min'].fillna(np.nanmedian(data_type1['hemoglobin_min']))
data_type1.info()
```

## include the lactate

```python
feature=''
for column in data_type1.drop(['outcome_group','age_group_0'],axis=1).columns.values:
    feature=feature+column+'+'
    
    
model = sm.GLM.from_formula("outcome_group ~ {}".format(feature[:-1]), family=sm.families.Binomial(), data=data_type1)
result = model.fit()
result.summary()
```

```python
result_df=pd.DataFrame(columns=['coef','std_err','p','[0.025','0.975]'])
result_df['coef']=result.params
result_df['std_err']=result.bse
result_df['p']=result.pvalues
result_df['[0.025']=result_df['coef']-1.96*result_df['std_err']
result_df['0.975]']=result_df['coef']+1.96*result_df['std_err']
result_df['OR']=np.exp(result_df['coef'])
result_df['floor']=np.exp(result_df['[0.025'])
result_df['upper']=np.exp(result_df['0.975]'])
result_df['p']=['%.2f'%x for x in result_df['p']]
result_df['OR']=['%.2f'%x for x in result_df['OR']]
result_df['floor']=['%.2f'%x for x in result_df['floor']]
result_df['upper']=['%.2f'%x for x in result_df['upper']]
for item in result_df.index:
    result_df.loc[item,'OR_out']=str(result_df.loc[item,'OR'])+' ({}-{})'.format(str(result_df.loc[item,'floor']),str(result_df.loc[item,'upper']))

result_df
```

```python
result_df.to_csv('../report/LR_result_include lactate.csv')
```
