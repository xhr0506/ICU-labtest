```python
import pandas as pd
import numpy as np
```

```python
path='F:\\项目\\301icu\\Leo文章\\extend information\\select_cohort_eicu.csv'
f=open(path, encoding='gbk',errors='ignore')
data_eicu=pd.read_csv(f)
```

```python
data_eicu.head()
```

```python
path='F:\\项目\\301icu\\Leo文章\\extend information\\eicu_crrt.csv'
f=open(path, encoding='gbk',errors='ignore')
eicu_extend_crrt=pd.read_csv(f)

path='F:\\项目\\301icu\\Leo文章\\extend information\\eicu_pivoted_vent.csv'
f=open(path, encoding='gbk',errors='ignore')
eicu_extend_vent=pd.read_csv(f)
```

```python
eicu_extend_vent['vent_duration_hours']=(eicu_extend_vent['endtime']-eicu_extend_vent['starttime'])/60
eicu_extend_vent.drop(['starttime','endtime'],axis=1,inplace=True)
eicu_extend_vent
```

```python
grouped=eicu_extend_vent.groupby(['patientunitstayid'])
eicu_extend_vent=grouped.sum()
eicu_extend_vent.reset_index(inplace=True)
eicu_extend_vent
```

```python
data_eicu=pd.merge(data_eicu,eicu_extend_crrt,on='patientunitstayid',how='left')
data_eicu=pd.merge(data_eicu,eicu_extend_vent,on='patientunitstayid',how='left')
data_eicu
```
