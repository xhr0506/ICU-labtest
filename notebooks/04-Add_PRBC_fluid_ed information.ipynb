## Read the updated tables

```python
data_update_mimic = pd.read_csv('E:/MyProject/ICU-labtest-range-global-study/datasets/MIMIC-data-update202012.csv',index_col=0)
data_update_aumc = pd.read_csv('E:/MyProject/ICU-labtest-range-global-study/datasets/AUMC-data-updated202012.csv',index_col=0)
data_update_eicu = pd.read_csv('E:/MyProject/ICU-labtest-range-global-study/datasets/eICU-data-update202012.csv',index_col=0)
data_update_plagh = pd.read_csv('E:/MyProject/ICU-labtest-range-global-study/datasets/PLAGH-data-update202012.csv',index_col=0)
data_update_HJ23 = pd.read_csv('E:/MyProject/ICU-labtest-range-global-study/datasets/HJ23-data-updated202012.csv',index_col=0)
```

```python
data_update_mimic.head(5)
```

```python
data_update_eicu.head(5)
```

```python
lower_col=[]
for col in data_update_plagh.columns.values:
    lower_col.append(str.lower(col))
data_update_plagh.columns=lower_col
data_update_plagh.head(5)
```

```python
data_update_aumc.head(5)
```

```python
data_update_HJ23=data_update_HJ23.loc[:,['subject_id','hadm_id','icustay_id','blood','colloid','crystalloid_v','emergency']]
data_update_HJ23.head(5)
```

## Process and Merge

```python
data_mimic_up=pd.merge(data_mimic,data_update_mimic,on=['subject_id','hadm_id','icustay_id'],how='left')
data_mimic_up.loc[data_mimic_up['crystalloid_v']>0,'crystalloid']=1
data_mimic_up.loc[:,'vent_flag':]=data_mimic_up.loc[:,'vent_flag':].fillna(0)
data_mimic_up.loc[:,['vent_duration_hours', 'crrt_duration_hours','crystalloid_v']]=data_mimic_up.loc[:,['vent_duration_hours', 'crrt_duration_hours','crystalloid_v']].replace(0,np.nan)
data_mimic_up
```

```python
data_mimic_up['crystalloid_v'].describe()
```

```python
data_eicu_up=pd.merge(data_eicu,data_update_eicu,on=['patientunitstayid'],how='left')
data_eicu_up.loc[(data_eicu_up['admission_type']=='EMERGENCY') & (data_eicu_up['unitadmitsource']=='Emergency Department'),'emergency']=1
data_eicu_up['emergency'].fillna(0,inplace=True)
data_eicu_up.loc[data_eicu_up['crystalloid_v']>0,'crystalloid']=1
data_eicu_up.loc[:,['vent_flag','rrt_flag', 'crystalloid_v','colloid','blood','crystalloid']]=data_eicu_up.loc[:,['vent_flag','rrt_flag', 'crystalloid_v','colloid','blood','crystalloid']].fillna(0)
data_eicu_up.loc[:,['vent_duration_hours', 'crrt_duration_hours','crystalloid_v']]=data_eicu_up.loc[:,['vent_duration_hours', 'crrt_duration_hours','crystalloid_v']].replace(0,np.nan)
data_eicu_up
```

```python
data_eicu_up['crystalloid_v'].describe()
```

```python
data_plagh_up=pd.merge(data_plagh,data_update_plagh,on=['patient_id','visit_id'],how='left')
data_plagh_up.loc[data_plagh_up['crystalloid_v']>0,'crystalloid']=1
data_plagh_up.loc[:,['colloid','blood','crystalloid']]=data_plagh_up.loc[:,['colloid','blood','crystalloid']].fillna(0)
data_plagh_up['crystalloid_v'].replace(0,np.nan,inplace=True)
data_plagh_up['vent_flag']=np.nan
data_plagh_up['rrt_flag']=np.nan
data_plagh_up['vent_duration_hours']=np.nan
data_plagh_up['crrt_duration_hours']=np.nan
data_plagh_up['cardiac_surgery_flag'].fillna(0,inplace=True)
data_plagh_up.loc[data_plagh_up['colloid']>=1,'colloid']=1
data_plagh_up
```

```python
data_plagh_up['crystalloid_v'].describe()
```

```python
data_aumc_up=pd.merge(data_aumc,data_update_aumc,on=['patientid','admissionid'],how='left')
data_aumc_up['emergency'].replace('ED',1,inplace=True)
data_aumc_up['emergency'].replace('Other',0,inplace=True)
data_aumc_up.loc[data_aumc_up['crystalloid_v']>0,'crystalloid']=1
data_aumc_up.loc[:,['vent_flag','crrt_flag','crystalloid_v','colloid','blood','crystalloid']]=data_aumc_up.loc[:,['vent_flag','crrt_flag','crystalloid_v','colloid','blood','crystalloid']].fillna(0)
data_aumc_up['crystalloid_v'].replace(0,np.nan,inplace=True)
data_aumc_up.rename(columns={'crrt_flag':'rrt_flag'},inplace=True)
data_aumc_up
```

```python
data_aumc_up['crystalloid_v'].describe()
```

```python
data_HJ23_up=pd.merge(data_HJ23,data_update_HJ23,on=['subject_id','hadm_id','icustay_id'],how='left')
data_HJ23_up['emergency'].replace('ED',1,inplace=True)
data_HJ23_up['emergency'].replace(['OR', 'Floor', 'OH'],0,inplace=True)
data_HJ23_up.loc[data_HJ23_up['crystalloid_v']>0,'crystalloid']=1
data_HJ23_up.loc[:,['vent_flag','rrt_flag','colloid','blood','crystalloid']]=data_HJ23_up.loc[:,['vent_flag','rrt_flag','colloid','blood','crystalloid']].fillna(0)
data_HJ23_up.loc[:,['vent_duration_hours', 'crrt_duration_hours', 'crystalloid_v']]=data_HJ23_up.loc[:,['vent_duration_hours', 'crrt_duration_hours', 'crystalloid_v']].replace(0,np.nan)
data_HJ23_up
```

```python
data_HJ23_up['crystalloid_v'].describe()
```
